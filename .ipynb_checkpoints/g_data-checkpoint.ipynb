```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from datetime import datetime
PATH = "/home/snaags/scripts/datasets/"
df = pd.read_csv(PATH+"g-research-crypto-forecasting/train.csv")
asset = pd.read_csv(PATH+"g-research-crypto-forecasting/asset_details.csv")
df["timestamp"] = df["timestamp"].apply(lambda x: datetime.fromtimestamp(x))
df["date"] = df["timestamp"].astype('datetime64[s]')
df.head()
df = pd.merge(df, asset, on="Asset_ID")
df.set_index("date", inplace=True)
df = df[["Asset_Name", "Weight", "Count",
         "Open", "High", "Low",
         "Close", "Volume", "VWAP",
         "Target"]]
df.isnull().mean() * 100
df.dropna(inplace=True)
df_btc = df.loc[df["Asset_Name"] == "Bitcoin"]
plt.plot(df_btc["Open"])
```

```python

df_btc = df_btc.loc[~df_btc.index.duplicated(), :]
```
